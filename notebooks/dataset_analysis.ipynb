```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
target_data = pd.read_csv("../data/dataset_stats.csv")
target_data.head()
```

```python
target_data['relations'].sum()
```

```python
sns.histplot(x='nodes', data=target_data)
plt.title("Distribuition of graphs nodes")
plt.show()
```

```python
sns.histplot(x='relations', data=target_data)
plt.title("Distribuition of graphs relations")
plt.show()
```

```python
target_data['mean_node_per_rel'] = target_data['nodes'] / target_data['relations']
```

```python
sns.histplot(x='mean_node_per_rel', data=target_data)
plt.title("Distribuition of graphs mean node per relation")
plt.show()
```
